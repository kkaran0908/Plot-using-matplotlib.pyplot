Matplotlib pyplot is the collection of command style function that make matplotlib work like matlab
Each pyplot function make some changes to a figure e.g. - plot a figure, creating a ploting area in a figure, plot some line in the ploting area, decorate the plot with some labels

```python
import matplotlib.pyplot as plt
plt.plot([2,4,6,4])
plt.ylabel("numbers")
plt.xlabel("Indices")
plt.title("my first plot")
plt.show()
```

If you provide a singlle list or array to the plot command, matplotlib assumes that it is the value of y axis and automatically generate the x values. Since the python range starts from 0, the default x vector has same length as y but starts with zero so x is [0,1,2,3]. If we provide the two dimensional list to the

```python
plt.plot([1,5,6,7],[0,3,6,7])
plt.ylabel("numbers")
plt.xlabel("Indices")
plt.title("second")
plt.show()
```

```python
plt.plot([1,2,3,4,5],[6,5,4,3,2],('ro'))
plt.ylabel("numbers")
plt.xlabel("indices")
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.,.2)
plt.plot(t,t**2,'b--',label ='2^')
plt.plot(t,t**2.2,'ro',label='2.2^')
plt.plot(t,t**2.8,'g',label='2.8^')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [4,5,6,7,7]
plt.plot(x,y,linewidth=5)
plt.grid()
plt.show()
```

```python
#plot multiple lines simulteneously given the end points of the lines.
x1 = [1,2,3,4,5]
y1 = [4,5,6,7,8]
x2 = [0,1,4,5,6]
y2 = [4,5,3,2,1]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='g',linewidth = 3)
plt.setp(lines[1],color='r',linewidth =1)
plt.grid()
plt.show()
```
